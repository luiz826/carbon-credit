# Emissões de $CO_2$

CO₂ emissions; CO₂ direct and indirect emissions per unit of output by industry and by country. CO₂ emissions by industry, in aggregate terms and in terms of output by industry

```python
df = pd.read_csv("./raw-data/emissoes_co2.csv")
```

```python
relevant_cols = [
    "Country",
    "Indicator",
    "Industry",
    "Unit",
]
relevant_cols.extend([f"F{i}" for i in range(1995, 2019)])

df_rel = df[relevant_cols]
```

```python
all_segments = df_rel[[
    "Country",
    "Indicator",
    "Industry",
    "Unit",
]]
```

```python
result = np.array([
    "Country",
    "Indicator",
    "Industry",
    "Unit",
    "Year",
    "Value"]).reshape(1,-1)

for row in df_rel.values:    
    segments = row[:4]
    vals = row[4:]
    n_years = len(vals)
    pivoted = np.c_[np.array([segments.tolist()]*n_years), np.arange(1995, 2019), vals]
    result = np.append(result, pivoted, axis=0)
```

```python
df_to_save = pd.DataFrame(result)
df_to_save.columns = df_to_save.iloc[0]
```

```python
df_to_save.iloc[1:].to_csv("./data/emissions.csv", index=False)
```

## Setor

```python
df5 = pd.read_csv("./raw-data/setor/4cn_1990-2016_agropecuaria.csv")
```

```python
ind = df5.iloc[:, 0][df5.iloc[:, 0] == "Ministério da Ciência, Tecnologia e Inovações"].index
```

```python
all_ind = [0] + ind.tolist()
for i in range(len(all_ind)-1):
    if i == 0:
        to_save = df5.iloc[all_ind[i]: all_ind[i+1]]
        to_save.to_csv(f"./data/setor/4cn_1990-2016_agropecuaria_{i+1}.csv", index=False)
    else:
        if i == len(all_ind)-1:
            to_save = df5.iloc[all_ind[i]+3: ]
        else:
            to_save = df5.iloc[all_ind[i]+3: all_ind[i+1]]
            
        to_save.columns = to_save.iloc[0]
        to_save.iloc[1:].to_csv(f"./data/setor/4cn_1990-2016_agropecuaria_{i+1}.csv", index=False)
        
    
```
